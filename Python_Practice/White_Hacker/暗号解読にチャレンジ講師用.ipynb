### ・チャプター1: 本日使用する関数と鍵の用意

```python
#文字と数字同士を変換する関数
import binascii

def int_to_str(m):
    hex_str = format(m, 'x')  
    if len(hex_str) % 2 != 0:
        hex_str = '0' + hex_str  
    byte_data = binascii.unhexlify(hex_str)
    try:
        return byte_data.decode('utf-8')    
    except UnicodeDecodeError:
        print("error")
        return None
        
def str_to_int(m):
    m_bytes = m.encode('utf-8')                       
    m_hex = binascii.hexlify(m_bytes)                 
    m_int = int(m_hex, 16)                           
    return m_int
```

```python
#使用例
message = "おはようございます"
massage_int = str_to_int(message)
print(f"文字に対応する数字(asciiコードと言います): {massage_int}")
message_str = int_to_str(massage_int)
print(f"数字を元の文字に戻した結果: {message_str}")
```

```python
#RSA暗号文の作成と解読を行う関数
"""
暗号化を行う関数: rsaEnc(数字に変換したメッセージ, 送信相手の公開鍵e, 送信する相手の公開鍵n)

暗号の解読を行う関数: rsaDec(暗号文、受け取り側の秘密鍵d, 受け取り側の公開鍵n)
"""

def rsaEnc(m, e, n):
    c = pow(m, e, n)
    return c

def rsaDec(c, d, n):
    m = pow(c, d, n)
    return m
```
